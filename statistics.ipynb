Q1. What is the difference between descriptive statistics and inferential
statistics? Explain with examples.

Ans.(a) **Descriptive Statistics** :

Descriptive statistics summarize and describe the main features of a dataset. It does not make predictions or generalizations beyond the given data.
Examples:

1. Mean salary of employees in a company.

2. Median marks of students in a class.

3. Bar chart showing monthly sales.

 **Tools Used:**
Mean, Median, Mode, Standard Deviation, Variance, Charts, Tables.

(b) **Inferential Statistics** :

Inferential statistics uses sample data to make conclusions or predictions about a larger population.
Examples:

1. Predicting election results using a sample survey.

2. Estimating average income of a city using 500 sampled households.

3. Hypothesis testing (e.g., testing if a new drug is effective).

**Tools Used:**
Confidence intervals, Hypothesis testing, Regression analysis, Correlation.


Q2.  What is sampling in statistics? Explain the differences between random
and stratified sampling ?


**Sampling**

Sampling is the process of selecting a subset of individuals from a population to estimate characteristics of the whole population.

(a) Random Sampling
- Every individual has an equal chance of being selected.
- Example: Picking 100 employees randomly from a list of 1000 employees.
- Advantages: Unbiased, Simple to implement

(b) Stratified Sampling
- Population is divided into subgroups (strata), and samples are taken from each group.
- Example: Dividing employees by department (HR, IT, Sales) and selecting samples from each department.
- Advantages: More representative, Better accuracy if groups differ



Q3.  Define mean, median, and mode. Explain why these measures of central

Ans.(a) **Mean** : Average value of data
	  
  (b)**Median**: Middle value when data is arranged in order.

Q4. Explain skewness and kurtosis. What does a positive skew imply about the
data?


Ans. **Skewness** : Measures asymmetry of data distribution.
- Positive Skew (Right Skew):
Tail is longer on right side.
Mean > Median.

- High kurtosis: More outliers.
- Low kurtosis: Flat distribution.

Positive Skew Implies about the data is that Most values are concentrated on the left with a few very high values pulling the mean upward.

Q5.  Implement a Python program to compute the mean, median, and mode of
a given list of numbers.
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

```python
import statistics
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

mean_value = statistics.mean(numbers)
median_value = statistics.median(numbers)
mode_value = statistics.mode(numbers)

print("Mean:", mean_value)
print("Median:", median_value)
print("Mode:", mode_value)
```

Q6. Compute the covariance and correlation coefficient between the following
two datasets provided as lists in Python:

list_x = [10, 20, 30, 40, 50]

list_y = [15, 25, 35, 45, 60]

```python
import numpy as np

list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

covariance = np.cov(list_x, list_y)[0][1]
correlation = np.corrcoef(list_x, list_y)[0][1]

print("Covariance:", covariance)
print("Correlation Coefficient:", correlation)
```

Conclusion :
- The covariance (275) indicates a positive relationship.
- The correlation coefficient (0.9907) indicates a very strong positive linear relationship.
- This means both variables move together closely.

Q8.  You are working as a data analyst in an e-commerce company. The
marketing team wants to know if there is a relationship between advertising spend and
daily sales.
- Explain how you would use covariance and correlation to explore this
relationship.
- Write Python code to compute the correlation between the two lists:
  - advertising_spend = [200, 250, 300, 400, 500]
  - daily_sales = [2200, 2450, 2750, 3200, 4000]


Ans. Using Covariance & Correlation

- Covariance shows direction (positive or negative).

- Correlation shows strength (-1 to +1).

- If correlation is close to +1 → strong positive relationship.

Interpretation:

Strong positive relationship — higher advertising spend increases sales.

```python
import numpy as np

advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

correlation = np.corrcoef(advertising_spend, daily_sales)[0][1]

print("Correlation:", correlation)
```

Q9 Your team has collected customer satisfaction survey data on a scale of
1-10 and wants to understand its distribution before launching a new product.
- Explain which summary statistics and visualizations (e.g. mean, standard
deviation, histogram) you'd use.
- Write Python code to create a histogram using Matplotlib for the survey data:
survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

Ans. **Summary Statistics to Use**
- Mean - Average satisfaction
- Median - Middle score
- Standard Deviation - Spread
- Min & Max - Range

**Visualization**
- Histogram -  Distribution pattern
- Boxplot -  Outliers

```python
import matplotlib.pyplot as plt

survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

plt.hist(survey_scores, bins=5)
plt.title("Customer Satisfaction Scores")
plt.xlabel("Scores")
plt.ylabel("Frequency")
plt.show()
```

**Interpretation:**
- Most scores are between 7–9.
- Distribution slightly left-skewed.
- Customers generally satisfied.
